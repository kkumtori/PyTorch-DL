# LIST
* 다양한 객체 유형을 담을 수 있는 순서가 정해진 Sequence.
* `[1,2,3,4,5]`
  * ,(콤마)를 이용해서 데이터를 구분


```python
my_list = [1,2,3]
my_list
```

```python
my_list = ['string', 100, 23.22, True, {1,2}] # 다양한 객체 유형 가능
my_list
```

```python
len(my_list) # 길이 측정(몇 개가 들어 있는지 알 수 있음.)
```

* indexing과 slicing 작업은 string형과 동일

```python
# indexing
my_list[0]
```

```python
# slicing
my_list[1:]
```

```python
another_list = ['four', 'five']
my_list + another_list # 리스트 더하기 연산 다른 변수로 선언하지 않으면 저장되지 않음
```

```python
# 문자형과는 다르게 요소를 변경할 수 잇음.
my_list[0] = 0
my_list
```

## 리스트 추가
* append method
  * list의 가장 마지막에 추가



```python
my_list.append('six') # 추가
my_list
```

## 리스트 제거
* pop method
  * 가장 마지막을 제거
  * 반환값으로 기존의 가장 마지막 값을 가짐.
  * 인덱스 값을 주어서 특정 위치의 요소를 지울 수 있음.(`pop(0)`)

```python
my_list.pop()
```

```python
my_list # 제거된 것을 확인할 수 있음.
```

## 리스트 정렬

```python
new_list.sort() # in-place method로 값을 반환하지 않고 기존값을 바꿈.
new_list
```

```python
type(new_list.sort()) # Nonetpye 객체가 없다는 의미로 아무것도 반환하지 않는 함수나 매서드의 반환값이다.
```

```python
new_list.reverse() # 역순정렬
new_list
```

# 딕셔너리
* 객체를 지정하는 순서 없이 매핑하는 방식.
* Hash Table 구조를 이용하기 때문에 내부 자료를 탐색할 때 속도가 리스트보다 빠르다.
  * 집합도 Hash Table 구조이다.  
  [참고링크](https://mangkyu.tistory.com/102)
* 키(key)와 값(value)를 이용해서 저장한다.
  * index를 알지 않아도 바로 값을 탐색할 수 있다.
    * 키와 값만 알고 있으면 된다.
    * 값이 가장 효율적인 위치에 자동으로 들어가게됨.
  * 중괄호를 이용해서 저장.
  * 순서가 정해져있지 않음.
  * `{key1:'value1', 'key2':value2'}`
  * 리스트와 마찬가지로 데이터를 저장하는 데이터 종류들이 생각보다 유연하다.


```python
# Make a dictionary with {} and : to signify a key and a value
my_dict = {'key1':'value1','key2':'value2'}
```

```python
my_dict
```

```python
# Call values by their key
my_dict['key2']
```

* 딕셔너리안에 다양한 변수 넣고 확인해보기

```python
# 딕셔너리에 만은 변수 넣어보기
my_dict = {'key1':123,'key2':[12,23,33],'key3':['item0','item1','item2']}
```

```python
# Let's call items from the dictionary
my_dict['key3']
```

```python
my_dict['key3'][0]
```

```python
# Can then even call methods on that value
my_dict['key3'][0].upper()
```

```python
# 딕셔너리 안에 딕셔너리 안에 딕셔너리 구조
d = {'key1':{'nestkey':{'subnestkey':'value'}}}
```

```python
d['key1']
```

```python
d['key1']['nestkey']
```

```python
d['key1']['nestkey']['subnestkey']
```

```python
# 모든 키 찾기
my_dict.keys()
```

```python
# 모든 값 찾기
my_dict.values()
```

```python
# 모든 키와 vlaue 찾기
my_dict.items()
```

# 튜플(tuple)
* 리스트와 매우 유사한 구조로 순서가 있는 데이터 구조이다.
  * 인덱싱, 슬라이싱 가능
* 불가변성 : 튜플 내에 인덱스 위치가 할당이 되면 그 요소를 가져와서 다른 것에 다시 할당할 수 없다.
  * 이 점이 리스트와의 차이점
  * `append` 사용 불가.
  * 값 내용 추가 불가.
  * 안에 몇개가 들어 있는지 셀 수 있다.(`count`이용)
  * 그 값이 나온 첫번째의 인덱스 위치만 알 수 있다.
* 소괄호를 이용해서 선언한다.  
  * ` a = (1,2,3)`
* 많이 사용하진 않지만, 불변성이 필요할 때는 종종 사용한다.

```python
t = (1, 2, 3)
```

```python
type(t)
```

```python
t[0]
```

```python
t[-1]
```

```python
t = ('a', 'a', 'b')
```

```python
t.count('a') #  a 가 몇개 있는 세기
```

```python
t.index('a')
```

```python
t.index('b')
```

```python
t[0] = 'index' # 오류 발생. immunity
```

# 집합(set)
* 말그대로 수학적 정의의 집합
  * 고유한 원소들의 순서 없는 모임.
* 값이 중복되어 들어가면 이전에 있던 값이 있으므로 값이 들어가지 않고 새로 들어온것에 대해서만 들어간다.(고유값만 값을 가지게 됨)
  * 그리고 값이 들어갈 때 오름차순으로 생성됨.

* 딕셔너리와 같이 해시 테이블 구조이며, 순서가 필요없다. 마찬가지로 중괄호를 사용하기 때문에 헷갈리지 않도록 유의해야한다.
* 빈 집합 생성
  * `set()`
* 요소 추가
  * `add(value)`

```python
a = [3,1,2]
myset = set(a)
```

```python
myset
```

```python
myset.add(4)
myset
```
