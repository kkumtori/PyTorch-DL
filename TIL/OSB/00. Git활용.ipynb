# GIt 활용

## Issue란?
* 프로젝트의 작업, 개산 사항 및 버그를 추적하는 좋은 방법으로 프로젝트 기획, 새롭게 추가될 기능, 버그와 수정사항 모든 것을이 Issue이다. 
* issue의 입력은 마크다운 형식으로 진행.
  * 보통 #을 통해서 내용과 
  * [ ]을 이용해 해야할 것 사용. 
* 라벨을 사용하여
  * Hash Tag처럼 어떤 글인지 빠르게 파악할 수 있다.
    * 손쉽게 검색
  * 부가적인 내용, 그룹화, 상태 등을 관리할 수 있음.
  * 회사마다 다른 규칙 사용.
  * 보통 타입, 상태, 카테고리로 나눠서 표시해줌.
    * 타입 - 버그, 기능개선, 신기능, 논의사항 ..
    * 상태 - New, In Progress, In Review, Done ...
    * 카테고리 - Front-end, Back-end, UI ...
  * 라벨을 한번 사용하면, 보통 기업됨.

## Web Hook
* 이벤트가 발생하면, 다른곳으로 이벤트가 발생했다고 알림을 주는 기능.
* 보통 push, merge, issue가 발생했다고 알림을 줄 수 있음.
  * 이를 이용해서 slack에 알림을 줄 수 있음.
  * 채널을 따로 하나 만들어서 해당 채널에 incoming webhooks를 이용한다.

## Issue와 Board의 관계
* 보통가 Tag별로 Issue를 구분해서 볼 수 있음
* 기본적으로 Open과 Closed로 구분되어 있음.
* 신규 보드를 생성해 원하는 대로 순서대로 볼 수 잇으며, 순서와 원하는 태그를 선택해 사용한다.

## Milestones
* 일종의 '목표'
* 프로젝트 관리를 하기 위해서 매우 유용함.
* 원하는 기능을 마일스톤으로 선택해서 진행
  * 보통 버전을 마일스톤으로 한다.

* 마일스톤 생성
  * 제목, 내용, 시작과, 종료시점을 정해 생성.
  * Issue를 생성된 마일스톤으로 할당해서 사용할 수 있음.
  * 만약, 할당된 마일스톤을 issue로 open하고 closed하면, 진행률을 확인할 수 있다.
  * Issue들의 진척도를 확인하기가 매우 수월해짐.

* Board와 병행해서 사용할 수 있으며, 기한과 시작일은 필수가 아니다.

## MR
* Merge Request의 줄임말.
* 다른 브랜치의 변경 사항을 기존 브랜치의 반영하기 위한 요청
  * 새로운 코드를 짰기 때문에 진행하는 과정.
  * 코드를 수정한(변경사항이 포함된) 브랜치를 소스 브랜치라 함.
  * 코드가 추가될(변경사항이 반영될) 브랜치를 타켓 브랜치라 함.
  * MR은 소스 브랜치 -> 타켓 브랜치로 진행되는 과정.
* 합치기 전에 검수를 하여 합치기 나서 발생할 문제들을 방지함.
* MR을 하게 되는 이유는 진행하는 프로젝트를 각 기능 별로 분류하여 진행해야지 코드 진행이 편리기 때문이다.

* 코드 검수를 위한 과정이므로 본인이 승인하지 않고, 담당 개발자나, 사수가 리뷰를 진행함,
  * 최소 Approver의 수를 정할 수 있다.
  * 리뷰를 진행하는 사람은 Approver로 따로 등록해야함.

### MR을 하는 이유
* 코드 리뷰
  * 병합 전에 코드 리뷰를 진행
* Master 브랜치으 안정성 유지
  * 실제로 합치고 나서 Master 브랜치가 잘 동작하다가 합치는 요소에 의해서 생기는 다양한 문제들을 방지하기 위해서 사용.
* 보안
  * 해킹을 방지하기 위해서도 사용.
  * 사람에 의한 보안 이슈를 방지할 수 있음.

## 이슈와 연동해서 MR 생성.
* 실제로는 기존 or 새로 생성된 issue 창에서 Branch를 생성(orMR까지)하거나 
* 생성된 Branch에서 코드 수정 후 Merge Request 진행.
  * 만약 Approver가 승인하면, 자동으로 issue가 closed됨.

## CI
* Continuous Integration(지속적인통합)의 줄임말.
* 구성원들이 작성한 내용을 정기적으로 통합하는 것을 의미.
  * 소프트웨어가 거대해지고 복잡해지면서 팀 단위로 개발이 진행되고 분업 및 협업이 필수가 되면서 등장하게 됨.
* 자동으로 [빌드 - 테스트 - 머지]를 해주는 기능이다.
* 새로운 코드에 대한 변경사항이 꾸준히 자동으로 빌드 및 테스트 되어 여러 개발자가 동시에 작업할 때 인터럽트가 발생하는 것을 방지할 수 있음.
  * 분엽과 협업을 효율적으로 할 수 있음.
  * 코드 작성에만 집중하도록 도와줌.
  * 안정성 유지 및 시간 절약에 필수적임.

* gitlab의 가장 큰 장점 중 하나.
  * CI/CD가 매우 잘됨.
  * CI-Pipeline을 통해 확인할 수 있음.
  * `.gitlab-ci.yml`에서  CI 기능에 대한 구성/설정을 저장하여 관리할 수 있음.



* 구성요소
  * CI 서버
  * SCM(Source code Management)
    * 소스코드 형상 관리 시스템으로 git이 여기에 포함됨.
  * build Tool
    * 컴파일, 테스트, 정적 분석 등을 실시해 동작 가능한 소프트웨어를 생성하는 도구(ex. Maven)
    * 형상 관리 시스템에 잇는 소스코드를 가져와 컴파일 하여 실행 가능한 파일로 만드는 일련의 과정.
  * Test Tool
    * 작성된 테스트 코드에 따라 자동으로 테스트를 수행해주는 도구로 빌드 틀의 script에서 실행되며 Junit이 여기에 해당한다.
